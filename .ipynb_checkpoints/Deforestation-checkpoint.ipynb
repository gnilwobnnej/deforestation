```python
#rasterio is a Python library used to read, write, analyze, and manipulate raster (gridded) geospatial data
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score

#utility to load red and nir bands from the satelite imagery and return it as a array float 32
def load_band(path):
    with rasterio.open(path) as src:
        return src.read(1).astype('float32')

#loads the photos before and after
red_before = load_band("before_RED.tiff")
nir_before = load_band("before_NIR.tiff")

red_after = load_band("after_RED.tiff")
nir_after = load_band("after_NIR.tiff")
```

```python
#calculates the Normalized Difference Vegetation Index (NDVI), a number between -1 and 1 that shows how healthy vegetation is.
def compute_ndvi(nir, red):
    # add epsilon to avoid division by zero
    ndvi = (nir - red) / (nir + red + 1e-10)  
    #Then it clips the result to make sure the NDVI stays between -1 and 1.
    return np.clip(ndvi, -1, 1)

#This calculates how much healthy vegetation was present before (using NIR and Red bands from a satellite image).
ndvi_before = compute_ndvi(nir_before, red_before)
#This does the same but for the after image.
ndvi_after = compute_ndvi(nir_after, red_after)

#This finds the change in vegetation by subtracting the after-image NDVI from the before-image NDVI.
ndvi_diff = ndvi_before - ndvi_after
```

```python
#This sets a cutoff value. Any NDVI change greater than 0.2 will be considered significant loss of vegetation.
threshold = 0.2
#This creates a mask (a map of True/False values) where: True means that area likely experienced deforestation (NDVI dropped a lot).
#False means no major change.
deforestation_mask = ndvi_diff > threshold
```

```python
#Sets up a wide figure for 3 plots.
plt.figure(figsize=(15, 5))

#Shows the vegetation before (from NIR + RED bands).Green areas = more healthy vegetation.
plt.subplot(1, 3, 1)
plt.title("NDVI Before")
plt.imshow(ndvi_before, cmap='Greens')
plt.colorbar()


#Shows vegetation after some event (e.g. deforestation).Comparing this to the first helps spot changes.
plt.subplot(1, 3, 2)
plt.title("NDVI After")
plt.imshow(ndvi_after, cmap='Greens')
plt.colorbar()

#Shows areas where NDVI dropped significantly.Red = likely deforested areas.
plt.subplot(1, 3, 3)
plt.title("Deforestation Detected")
plt.imshow(deforestation_mask, cmap='Reds')
plt.colorbar()

#Makes sure plots don't overlap and shows them.
plt.tight_layout()
plt.show()
```

```python
#np.zeros_like(red_before): Makes a blank array the same shape as your RED band image.
#dtype='uint8': Data type = 8-bit unsigned integer (values from 0 to 255). Perfect for binary masks.
ground_truth = np.zeros_like(red_before, dtype='uint8')

#This marks a rectangular region from row 50 to 99 and column 100 to 149 as 1, which we treat as "deforested".
#manually labeling a region you know was cleared in the image.
ground_truth[50:100, 100:150] = 1

#Opens a new file called deforestation_truth.tiff for writing.
#Makes sure it has the same size and geolocation as your original image.
with rasterio.open(
    'deforestation_truth.tiff',
    'w',
    driver='GTiff',
    height=ground_truth.shape[0],
    width=ground_truth.shape[1],
    count=1,
    dtype='uint8',
    crs='+proj=latlong',
    transform=rasterio.open("before_RED.tiff").transform  # reuse georeferencing
) as dst:
    dst.write(ground_truth, 1)
```

```python
#load_band- Loads the .tiff file where you manually marked deforested areas (0 = not deforested, 1 = deforested).
#0.5: Ensures the values are converted to True (1) and False (0). This makes it a boolean mask, ready for evaluation.
truth_mask = load_band("deforestation_truth.tiff")
truth_mask = truth_mask > 0.5

#Prevents errors later on by confirming that both masks (your prediction and the true data) are the same size.
if deforestation_mask.shape != truth_mask.shape:
    raise ValueError("Shape mismatch between predicted and ground truth masks.")

#flattens the image and turns a 2d image into a 1d array
y_true = truth_mask.flatten()
y_pred = deforestation_mask.flatten()

#Compares your models predictions (y_pred) with the actual labels (y_true) and calculates the F1 score:
f1 = f1_score(y_true, y_pred)
print(f"F1 Score for Deforestation Detection: {f1:.4f}")
```

```python
print("Predicted deforested pixels:", np.sum(deforestation_mask))
print("True deforested pixels:", np.sum(truth_mask))
```

```python
# Plot predicted vs ground truth deforestation masks
plt.figure(figsize=(12, 6))

# Prediction
plt.subplot(1, 2, 1)
plt.title("Predicted Deforestation Mask")
plt.imshow(deforestation_mask, cmap='Reds')
plt.colorbar()

# Ground truth
plt.subplot(1, 2, 2)
plt.title("Ground Truth Deforestation Mask")
plt.imshow(truth_mask, cmap='Reds')
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
# Visualize the NDVI difference map directly
plt.imshow(ndvi_diff, cmap='RdBu', vmin=-1, vmax=1)
plt.title("NDVI Difference Map (Before - After)")
plt.colorbar()
plt.show()
```

IoU (Intersection over Union) measures how much your prediction overlaps with the ground truth.

Intersection = the pixels both prediction and truth say are deforested (true positives).

Union = all the pixels that either prediction or truth say are deforested.

overlapping = 0 no overlap 1 perfect match

```python
#Creates a mask of pixels where either says it's deforested (so both true positives and false positives/negatives).
intersection = np.logical_and(deforestation_mask, truth_mask)
union = np.logical_or(deforestation_mask, truth_mask)

#Counts the number of pixels where prediction and ground truth overlap, and divides it by the total number of pixels flagged by either.
iou = np.sum(intersection) / np.sum(union)
print(f"IoU (Intersection over Union): {iou:.4f}")
```

```python
#What it does: Finds areas where vegetation dropped a lot  more than 0.25 in NDVI value.
#This makes sure we only call something "deforested" if the change is big enough. It avoids calling small changes deforestation.
truth_mask = ndvi_diff > 0.25  

#Opens the original RED band file to copy its geographic info (like size, location on Earth).
ground_truth = truth_mask.astype('uint8')

# Save it with the same geo-referencing as before_RED
with rasterio.open("before_RED.tiff") as ref:
    profile = ref.profile
    profile.update(dtype='uint8', count=1)

    with rasterio.open("deforestation_truth_auto.tiff", 'w', **profile) as dst:
        dst.write(ground_truth, 1)
```

```python
#Load and threshold
truth_mask = load_band("deforestation_truth_auto.tiff") > 0.5

#flattens 2d to 1d array vec
y_true = truth_mask.flatten()
y_pred = deforestation_mask.flatten()

#Compares your predicted mask vs the new truth mask.
f1 = f1_score(y_true, y_pred)
print(f"F1 Score with aligned ground truth: {f1:.4f}")
```
